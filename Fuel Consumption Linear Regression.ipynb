```python
## Its time to create testing and training datasets
```

```python
X = fuel_economy_df[['Horse Power']]
```

```python
Y = fuel_economy_df[['Fuel Economy (MPG)']]
```

```python
X.shape
```

```python
Y.shape
```

```python
## ITS TIME FOR TRAINING
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30)
```

```python
X_train
```

```python
X_test
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(fit_intercept = True)
regressor.fit(X_train, Y_train)
```

```python
print('Linear Model Coeff (m):' , regressor.coef_)
print('Linear Model Coeff (b):' ,regressor.intercept_)
```

```python
y_predict = regressor.predict(X_test)
```

```python
y_predict
```

```python
plt.scatter(X_train, Y_train, color = 'green')
plt.plot(X_train, regressor.predict(X_train), color = 'black')
plt.xlabel('Horse Power(HP)')
plt.ylabel('(MPG) Fuel')
plt.title('MPG VS HP (TRAINING DATASET)')
```

```python
plt.scatter(X_test, Y_test, color = 'green')
plt.plot(X_test, regressor.predict(X_test), color = 'black')
plt.xlabel('Horse Power(HP)')
plt.ylabel('(MPG) Fuel')
plt.title('MPG VS HP (TESTING DATASET)')
```
